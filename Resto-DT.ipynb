```python
#drop all columns with null values
data.dropna(axis=1, inplace=True)
```

```python
sns.heatmap(data.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
data.head()
```

```python
data.drop(['Timestamp','nama','umur'], axis=1, inplace=True)
```

```python
data.head()
```

```python
data['Apakah menurutmu tempat makan saat ini sudah bisa dibuka kembali?'] = data['Apakah menurutmu tempat makan saat ini sudah bisa dibuka kembali?'].map({'Ya':1, 'Tidak':0})
data['Apakah prosedur kesehatan yang digunakan di tempat makan sudah efektif?'] = data['Apakah prosedur kesehatan yang digunakan di tempat makan sudah efektif?'].map({'Ya':1, 'Tidak':0})
data['Apakah menurutmu tempat makan saat ini sudah bisa dibuka kembali?'].fillna(1.0, inplace=True)
data['Apakah prosedur kesehatan yang digunakan di tempat makan sudah efektif?'].fillna(1.0, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
Y = data['Apakah menurutmu tempat makan saat ini sudah bisa dibuka kembali?']
X = data.drop(['Apakah menurutmu tempat makan saat ini sudah bisa dibuka kembali?'], axis=1)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier(random_state=101)
dtree.fit(X_train, Y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(Y_test,predictions))
```

```python
print(classification_report(Y_test,predictions))
```

```python
#model validation
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=dtree, X=X_train, y=Y_train, cv=10)
print(accuracies.mean())
print(accuracies.std())
```
